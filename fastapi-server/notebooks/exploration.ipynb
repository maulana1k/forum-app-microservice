```python
import pandas as pd
import numpy as np

df = pd.read_csv('../dataset/raw/chatgpt1.csv')

df.head()
```

```python
df.info()
```

```python
# List top posts by ReplyCount
top_posts = df.sort_values(by='ReplyCount', ascending=False)[['Text', 'ReplyCount']].head(10)
top_posts
```

```python
user_counts = df.groupby('Username')['Tweet Id'].count().sort_values(ascending=False)
user_ranks = user_counts.rank(method='dense', ascending=False).astype(int)
result = pd.DataFrame({'TweetCount': user_counts, 'Rank': user_ranks})
result.head()
```
